```python
def plot_data(string,y_scale=False):
    """
        Inputs : Name of the pkl file containing the history object.
        Outputs : A graph of training loss versus the epoch.
    """
    
    history = pkl.load(open(string,"rb"))
    print ("The data pertains to : ",string,"\n")
    
    #----------------------------------------------#
    plt.figure(figsize=(15,10))
    plt.grid(True)
    plt.title("Training Loss variations over the epochs.")
    plt.xlabel("Epoch Number")
    plt.ylabel("Training Loss")
    if(y_scale):
        plt.ylim([0,20])
    plt.plot(history["loss"])
    plt.legend(["Training Loss"])
    plt.show()
    #----------------------------------------------#
    
    return
```

```python
#NYC Load Data
plot_data("Results/M1_V1_NYC.pkl")
```

```python
#Stocks - Amazon.
plot_data("Results/M1_V1_AMZN.pkl")
```

```python
#Stocks - FB.
plot_data("Results/M1_V1_FB.pkl")
```

```python
#Stocks - FB.
plot_data("Results/M1_V1_WMT.pkl",True)
```
